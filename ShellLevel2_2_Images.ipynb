```python
## Images
def makeData(n):
    X = []
    Y = []

    for date in dates:
        part = data[data['DATE (YYYY/MM/DD)']==date].copy()
        part.reset_index(inplace = True, drop = True)
        part.drop(['DATE (YYYY/MM/DD)', 'MST'], axis=1, inplace=True)

        if part.shape[0] >= (n + 120):
            for i in range((part.shape[0]//10) - (int(n/10) + 11)):
                X.append(part.iloc[i*10:n+i*10].to_numpy())    ##########Covered n (250) blocks or rows but covered 249 minutes of data

                targets = []
                for j in range(1,13):
                    targets.append(part.iloc[n-1+j*10+i*10]['Global CMP22 (vent/cor) [W/m^2]'])

                Y.append(targets)
                    
        print(date)
    
    X = np.array(X)
    Y = np.array(Y)

    return X,Y
```

```python
def lstm(n):
    model = keras.Sequential()
    model.add(LSTM(512, input_shape=(n, 16), return_sequences = True)) #512
    # model.add(tf.keras.layers.Dropout(0.3))  # added
    model.add(LSTM(256, return_sequences = True))   #128
    model.add(LSTM(128))  #64

    model.add(tf.keras.layers.Dense(1024,kernel_regularizer=tf.keras.regularizers.L2(0.0001),activation = 'relu'))

    model.add(tf.keras.layers.Dense(512,kernel_initializer=initializers.RandomNormal(stddev=np.sqrt(2.0/1024.0)),kernel_regularizer=tf.keras.regularizers.L2(0.0001),activation = 'relu'))

    model.add(tf.keras.layers.BatchNormalization(
        axis=-1, momentum=0.99, epsilon=0.001, center=True, scale=True,
        beta_initializer='zeros', gamma_initializer='ones',
        moving_mean_initializer='zeros',
        moving_variance_initializer='ones', beta_regularizer=None,
        gamma_regularizer=None, beta_constraint=None, gamma_constraint=None,
        renorm=False, renorm_clipping=None, renorm_momentum=0.99, fused=None,
        trainable=True, virtual_batch_size=None, adjustment=None, name=None))
    #128 down
    model.add(tf.keras.layers.Dense(64,kernel_initializer=initializers.RandomNormal(stddev=np.sqrt(2.0/512.0)),kernel_regularizer=tf.keras.regularizers.L2(0.0001),activation = 'relu'))
    # model.add(tf.keras.layers.Dropout(0.2))  # added
    model.add(tf.keras.layers.Dense(64,kernel_initializer=initializers.RandomNormal(stddev=np.sqrt(2.0/512.0)),kernel_regularizer=tf.keras.regularizers.L2(0.0001),activation = 'relu'))  # added

    model.add(Dense(12,kernel_initializer=initializers.RandomNormal(stddev=np.sqrt(2.0/128.0))))

    return model
```

```python
model = lstm(250)
model_450 = lstm(450)
model_550 = lstm(550)
```

```python
initial_learning_rate = 0.001
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=1000,
    decay_rate=0.96,
    staircase=True)
optimizer = Adam(learning_rate=lr_schedule,beta_1=0.9, beta_2=0.999, epsilon=1e-08)
model.compile(optimizer=optimizer,
              loss=tf.keras.losses.Huber(delta=4.0),
              metrics=['mean_absolute_error'])
```
